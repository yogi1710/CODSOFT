Importing required libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the data set to a variable using pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# head() method returns top 5 data points from the dataset
credit_card_data.head()
```

```python
# tail() method retuens the last 5 data points from the data set
credit_card_data.tail()
```

```python
# info() method returns the information of the given dataset
credit_card_data.info()
```

```python
# isnull() returns 0 if data is null else 1 for individual values
# by adding all such values we get the total no.of null values - we use sum() method
credit_card_data.isnull().sum()
```

```python
# As the data is fully available, we don't need to fill the values
# Distribution of ligit and fraudalant transaction
credit_card_data["Class"].value_counts()
# The below data is unbalanced because ,more than 90% of data is in one particular class
# 0 ------> 284315(Normal Transaction)
# 1 ------> 492(Fraudalant Transation)
# Hence the data is unbalanced
```

```python
# Dividing the data for further analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Finding the shape of divided data
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measurements of the legit data
# like mean, min, max, std etc..
legit.Amount.describe()
```

```python
# Statistical measurements of the fraud data
# like mean, min, max, std etc..
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
# It groups the similar class and then find the mean of the grouped data set
credit_card_data.groupby('Class').mean()
```

```python
# Under sampling
# Building a sample dataset from original dataset thaat contains similar distributions of legit and fraud transactions
# Number of a fraudalant Transactions - 492
# From the legit transaction take 429 random sample data by using sample() method
legit_sample = legit.sample(n = 429)
```

```python
# Concating the two dataframes
# After getting the sample data concat the both fraud and legit sample dataset
new_data = pd.concat([legit_sample, fraud], axis = 0)
# if axis is 0, values by row
# if axis value is 1, value by column
```

```python
# head() of the new dataset
new_data.head()
```

```python
# counts the available values and dispaly them
new_data["Class"].value_counts()
# 0,1 --> 429
```

```python
# finds the mean and returns it
new_data.groupby("Class").mean()
```

```python
# Splitting the data into Features and Targets
# Drop the Class value
# storing the Class value in other variable separatly
X = new_data.drop(columns = "Class", axis = 1)
Y = new_data["Class"]
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
# 20% of the data is used for test data
# 80% of the data is used for train data
```

```python
# shape of X,X_test and X_trian
print(X.shape, X_train.shape, X_test.shape)
```

Model Training using Logidtic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model eith training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# Model evaluation using Accuracy Score
# Compares the predicted values from X_train_prediction and actual values from Y_train
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on the training data : ", training_accuracy)
```

```python
# Evaluating the model by using unseen data is test data
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuacy on the testing data : ", testing_accuracy)
```
